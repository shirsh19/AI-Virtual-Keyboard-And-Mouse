```python
import cv2
from cvzone.HandTrackingModule import HandDetector
import pyautogui  

cap = cv2.VideoCapture(0)
cap.set(3, 1280)  
cap.set(4, 720)   

detector = HandDetector(detectionCon=0.8, maxHands=2)  
keys = [["Q", "W", "E", "R", "T", "Y", "U", "I", "O", "P"],
        ["A", "S", "D", "F", "G", "H", "J", "K", "L", ";"],
        ["Z", "X", "C", "V", "B", "N", "M", ",", ".", "/"]]


def drawALL(img, buttonList):
    for button in buttonList:
        x, y = button.pos
        w, h = button.size
        cv2.rectangle(img, button.pos, (x+w, y+h), (255, 0, 255), cv2.FILLED)
        cv2.putText(img, button.text, (x+20, y+65), cv2.FONT_HERSHEY_PLAIN, 3, (255, 255, 255), 3)
    return img


class Button():
    def __init__(self, pos, text, size=[85, 85]):
        self.pos = pos
        self.size = size
        self.text = text

buttonList = []
for i in range(len(keys)):
    for j, key in enumerate(keys[i]):
        buttonList.append(Button([100*j + 50, 100*i + 50], key))


while True:
    success, img = cap.read() 
    if not success:
        break

    hands, img = detector.findHands(img)
    
    if len(hands) == 1:  
        hand = hands[0]  
        lmList = hand['lmList']  

        if lmList:
            index_finger = lmList[8][:2]  
            thumb_tip = lmList[4][:2]     

            screen_width, screen_height = pyautogui.size()
            cursor_x = int(index_finger[0] * screen_width / 1280)  
            cursor_y = int(index_finger[1] * screen_height / 720)
            pyautogui.moveTo(cursor_x, cursor_y)

            distance, _, _ = detector.findDistance(lmList[8][:2], lmList[4][:2])
            if distance < 30:  
                pyautogui.click()

            distance_drag, _, _ = detector.findDistance(lmList[8][:2], lmList[12][:2])
            if distance_drag < 30: 
                pyautogui.mouseDown()
            else:
                pyautogui.mouseUp()

    elif len(hands) == 2:  
        img = drawALL(img, buttonList)
        hand = hands[0]  
        lmList = hand['lmList']  

        if lmList:
            for button in buttonList:
                x, y = button.pos
                w, h = button.size

                if x < lmList[8][0] < x + w and y < lmList[8][1] < y + h:
                    cv2.rectangle(img, button.pos, (x + w, y + h), (175, 0, 175), cv2.FILLED)
                    cv2.putText(img, button.text, (x + 20, y + 65), cv2.FONT_HERSHEY_PLAIN, 3, (255, 255, 255), 3)

                    l, _, _ = detector.findDistance(lmList[8][:2], lmList[12][:2])
                    if l < 30:
                        pyautogui.write(button.text)  
                        cv2.rectangle(img, button.pos, (x + w, y + h), (0, 255, 0), cv2.FILLED)
                        cv2.putText(img, button.text, (x + 20, y + 65), cv2.FONT_HERSHEY_PLAIN, 3, (255, 255, 255), 3)

    cv2.imshow("Image", img)

    if cv2.waitKey(1) & 0xFF == ord('Q'):
        break

cap.release()
cv2.destroyAllWindows()
```
